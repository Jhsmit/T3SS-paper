```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from colicoords.preprocess import data_to_cells, filter_binaries
from colicoords import Data, iter_subplots, CellListPlot, IterCellPlot, AutoIterCellPlot, save
import os
import mahotas as mh
```

```python
import colicoords
colicoords.__file__
```

```python
#input_dir = r'F:\_processed_big\t3ss\20191002_deltaescv_c41_eyfp-escv_repeat_03'
input_dir = r'H:\20191002_yichen_deltaescv_c41_eyfp_escv_repeat_04'
```

```python
c41_bf = np.load(os.path.join(input_dir, 'BF_C41_flat.npy'))
c41_bin = np.load(os.path.join(input_dir, 'c41_segmented.npy'))
c41_g100 = np.load(os.path.join(input_dir, 'c41_514_g100.npy'))
c41_g500 = np.load(os.path.join(input_dir, 'c41_514_g500.npy'))
```

```python
c41_g500.shape
```

```python
c41_b_filter = filter_binaries(c41_bin, min_size=200, max_size=10000, min_minor=6, max_minor=11, min_major=10, max_major=100)
```

```python
data = Data()
data.add_data(c41_b_filter, 'binary')
data.add_data(c41_bf, 'brightfield')
data.add_data(c41_g100, 'fluorescence', 'g100')
data.add_data(c41_g500, 'fluorescence', 'g500')
```

```python
cells = data_to_cells(data, remove_multiple_cells=False)
```

```python
fig, (ax1, ax2) = iter_subplots(1, 2)
ax1.iter_imshow(c41_bf)
ax2.iter_imshow(c41_b_filter)
fig.display()
```

```python
len(cells)
```

```python
aicp = AutoIterCellPlot(cells)
aicp.plot()
```

```python
save('c41_cell_raw.hdf5', cells)
```
